```python
# Renombrar la columna 'Power' eliminando espacios en blanco al inicio y al final
data.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
data.rename(columns={'Etching Rate': 'Etching_Rate'}, inplace=True)
```

```python
#calular la media y agrupar por "power"
data.groupby('Power').agg(np.mean)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Etching_Rate ~ C(Power)', data=data).fit()
sm.stats.anova_lm(model)

#un valor p bajo (en este caso, 0.000008), generalmente menor que 0.05, sugiere que la diferencia es estadísticamente significativa por lo que los resultados indican que existe una relación significativa entre la variable "Power" y la variable "Etching Rate" 
```

```python
from scipy.stats import ttest_ind

display_a = data[data['Power'] == '200']['Etching_Rate']

for display in data['Power'].unique():
    display_b = data[data['Power'] == display]['Etching_Rate']
    print(display, ttest_ind(display_a, display_b))
```
